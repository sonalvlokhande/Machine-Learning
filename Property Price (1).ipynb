# Sampling

# Building the model 

```python
from sklearn.linear_model import LinearRegression 
```

```python
linereg = LinearRegression()
```

```python
linereg.fit(pp_train_x,pp_train_y)
```

```python
#Rsquare
Rsquare = linereg.score(pp_train_x,pp_train_y)
Rsquare
```

```python
N = pp_train_x.shape[0]
K = pp_train_x.shape[1]
```

```python
#adj_Rsquare
adj_Rsquare = 1-(1-Rsquare)*(N-1) / (N-K-1)
adj_Rsquare
```

```python
pred_train = linereg.predict(pp_train_x)
```

```python
pred_test = linereg.predict(pp_test_x)
```

```python
err_test = pp_test_y - pred_test
err_test
```

```python
#mse
mse = np.mean(np.square(err_test))
mse
```
